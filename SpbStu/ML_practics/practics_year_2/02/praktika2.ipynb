```python
import requests
import json
```

```python
url = "https://suggestions.dadata.ru/suggestions/api/4_1/rs/suggest/party"
api_key = "0e575e6fe86b527a23910bf0bddb3cc173a658f3"  
headers = {
        "Content-Type": "application/json",
        "Accept": "application/json",
        "Authorization": f"Token {api_key}"
        }
```

```python
query = "Сбербанк"
count = 20
payload = {
        "query": query,
        "count": count  
}
try:
    response = requests.post(url, headers=headers, json=payload)
    response.raise_for_status()  # Проверка на ошибки HTTP
        
    data = response.json()
    organizations = data.get("suggestions", [])
except requests.exceptions.RequestException as e:
    print(f"Ошибка при выполнении запроса: {e}")
    organizations = []
```

```python
if not organizations:
    print("Организации не найдены.")
print(f"Найдено организаций: {len(organizations)}\n")
for org in organizations:
    data = org.get("data", {})
    name = data.get("name", {}).get("full_with_opf", "Н/Д")
    inn = data.get("inn", "Н/Д")
    address = data.get("address", {}).get("value", "Н/Д")
    status = data.get("state", {}).get("status", "Н/Д")
    manager_name = data.get("management", {}).get("name", "Н/Д") if data.get("management") else "Н/Д"
        
    # Преобразование статуса в читаемый формат
    status_map = {
        "ACTIVE": "действующая",
        "LIQUIDATED": "ликвидированная",
        "LIQUIDATING": "ликвидируется",
        "REORGANIZING": "в процессе реорганизации"
    }
    status_text = status_map.get(status, status)
        
    print(f"Полное название: {name}")
    print(f"ИНН: {inn}")
    print(f"Адрес: {address}")
    print(f"Статус: {status_text}")
    print(f"ФИО руководителя: {manager_name}")
    print("-" * 50)
```
