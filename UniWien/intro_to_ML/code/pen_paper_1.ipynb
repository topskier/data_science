```python
import numpy as np
```

```python
X = np.array([[1,1,1],[2,2,0],[3,3,-1]])
y = np.array([[1],[2],[3]])
print(X)

X_t_X= np.dot(X.T,X)
print(X_t_X)
X_t_y=np.dot(X.T,y)
print(X_t_y)
```

```python
X = np.array([[1,1],[2,0],[3,-1]])
y = np.array([[1],[2],[3]])
print(X)

X_t_X= np.dot(X.T,X)
print(X_t_X)
X_t_y=np.dot(X.T,y)
print(X_t_y)
print(np.linalg.inv(X_t_X))
w=np.dot(np.linalg.inv(X_t_X),X_t_y)
print(w)
```
